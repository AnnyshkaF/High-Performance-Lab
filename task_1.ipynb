```python
#Bifurcation map

import matplotlib.pyplot as plt
import numpy as np
from random import random
```

```python
# IMPLEMET THE MAP
n = 300 
r = 4

x = random()
X = []

def mapping(r, x):
    return r*x*(1-x)

for i in range (n):
    x = mapping(r, x)
    X.append(x)
        

# plot the evolution of X
plt.suptitle('Evolution of x for r = {}'.format(r))
plt.xlabel('Iterations after n = {}'.format(n))
plt.ylabel('X')
plt.plot(X)
```

```python
# Create a linspace of rs, for every r 
# save last m values of x after first
# n values (can be m=200, x=200) 
# play around with values

def get_bifurcation_map(n, m, r_beg, r_end, r_step):
  x = random()  
  r_values = np.linspace(r_beg, r_end, r_step)

  res = []
  for r in r_values:
      X = []
      x = random()
      for i in range(n+m):
          x = mapping(r, x)
          if(i > n):
            X.append(x) 

      res.append(X)

  # Get the bifurcation map
  plt.figure(figsize=(20, 5))
  plt.xlabel('r')
  plt.ylabel('X')
  _ = plt.plot(r_values, res, 'b.', markersize = 0.1)

get_bifurcation_map(200,200,0,5,1000)
get_bifurcation_map(200,30,0,5,1000)
```
